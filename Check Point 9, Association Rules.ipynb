```python
import mlxtend
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules 
from mlxtend.frequent_patterns import association_rules 
```

# 1st Dataset

```python
smaller_data = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

    ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

    ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

    ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

   ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
te=TransactionEncoder()
te_ary=te.fit(smaller_data).transform(smaller_data)    #Apply one-hot-encoding on our dataset
ds=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
ds
```

```python
apriori(ds, min_support=0.6)
```

```python
frequent_itemsets=apriori(ds, min_support=0.6, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

```python
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7) # associate itemsets with confidence over 70%.
```

```python
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```

```python
dataset2=[]
for i in range(0,7500):
    dataset2.append([str(bigger_data.values[i,j]) for j in range(0,20)])
```

```python
te2= TransactionEncoder()
te_ary2 = te2.fit(dataset2).transform(dataset2)
df2=pd.DataFrame(te_ary2,columns=te2.columns_)
df2
```

```python
ds2=pd.DataFrame(df2.astype("int").sum(),columns=['count'])
ds2
```

```python
freq_itemsets2=apriori(df2, min_support=0.03,use_colnames=True, verbose=1)
freq_itemsets2.sort_values(by="support").reset_index(drop=True)
```

```python
apriori(df2, min_support=0.6)
```

```python
frequent_itemsets2=apriori(df2, min_support=0.6, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets2
```

```python
association_rules(freq_itemsets2,metric="confidence",min_threshold=0.1).sort_values(by='confidence',ascending=False).reset_index(drop=True)
```

```python
association_rules(freq_itemsets2,metric="lift",min_threshold=1).sort_values(by='lift',ascending=False).reset_index(drop=True)
```
